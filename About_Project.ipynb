<b>Crew AI</b> is an open-source framework that enables you to build multi-agent AI systems where multiple AI agents collaborate to perform complex tasks. It allows you to define individual agents with specific roles, reasoning abilities, and tools, and then coordinate them within a structured workflow.

##### <b> Key Components of Crew AI </b>

1. Agent  Define multiple AI agents with unique roles and objectives.

2. Task - Assign tasks and dependencies to agents for structured workflows.

3. Tool  Agents can use APIs, databases, and external tools to enhance capabilities.

4. Memory  Agents can retain knowledge across interactions.

5. Parallel  Supports multiple agents working simultaneously.
